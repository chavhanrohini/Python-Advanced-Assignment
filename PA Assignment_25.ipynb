Q1. What is the distinction between a numpy array and a pandas data frame? Is there a way to convert between the two if there is?

- Numpy Array: Numpy arrays are homogeneous, multidimensional arrays that store data of the same data type. They are primarily used for numerical operations and data manipulation. Numpy provides efficient numerical operations and functions for arrays.

- Pandas DataFrame: A pandas DataFrame is a two-dimensional, heterogeneous tabular data structure that can store data of different data types. It is built on top of numpy arrays but provides more flexibility and functionality for data analysis, manipulation, and exploration.

You can convert a pandas DataFrame to a numpy array using the `.values` attribute of the DataFrame. For example:

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Convert DataFrame to numpy array
numpy_array = df.values
```

Q3. Identify some of the plotting techniques that are used to produce a stock-market chart.

Stock market charts can be created using various plotting techniques, including:
- Line Charts: Displaying the price or performance of a stock over time.
- Candlestick Charts: Providing information about open, close, high, and low prices in a given time period.
- Bar Charts: Displaying price changes and volumes for each time period.
- OHLC (Open, High, Low, Close) Charts: Combining elements of candlestick and bar charts to represent stock data.
- Moving Averages: Plotting moving averages to visualize trends and patterns.
- Bollinger Bands: Indicating volatility and potential reversal points.
- Volume Charts: Showing trading volume alongside price data.

Q6. What is the definition of a 180-day moving average?

A 180-day moving average is a technical indicator used in financial analysis and stock market charting. It represents the average closing price of a stock or asset over the past 180 trading days. To calculate it, you sum the closing prices of the asset for the most recent 180 days and divide by 180.

The 180-day moving average is often used to smooth out short-term fluctuations in stock prices and reveal longer-term trends. Traders and investors use it to identify potential trend reversals or confirm existing trends.

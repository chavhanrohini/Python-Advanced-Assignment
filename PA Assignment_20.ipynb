1. Compare and contrast the float and Decimal classes' benefits and drawbacks.

Floats:
- Benefits:
  - Floating-point numbers are typically faster to perform arithmetic operations on due to hardware support.
  - Floats can represent a wide range of values, both very small and very large.
- Drawbacks:
  - Limited precision: Floats cannot represent all real numbers exactly, which can lead to rounding errors.
  - Accuracy issues: Floats can't always accurately represent decimal numbers, especially when dealing with fractions.
  - Loss of precision in calculations.

Decimal:
- Benefits:
  - Arbitrary precision: Decimal class in Python provides arbitrary precision, allowing you to represent and perform precise arithmetic on decimal numbers.
  - Control over rounding: You can specify rounding rules, which can help avoid unexpected rounding errors.
- Drawbacks:
  - Slower than floats: Decimal arithmetic is generally slower due to the extra precision and operations involved.
  - Larger memory usage: Decimal objects consume more memory compared to floats.

7. Using the Fraction class but not the Decimal class, give an example of a quantity that can be expressed with absolute precision.

An example of a quantity that can be expressed with absolute precision using the `Fraction` class is 1/3. When you represent 1/3 as a Fraction, it will be exactly represented without any rounding or loss of precision.
